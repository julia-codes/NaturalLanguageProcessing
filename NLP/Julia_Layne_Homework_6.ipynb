```python
#Building a matrix for comparisons
for book in books.BookTitle:
    books[book] = '-'
books.head(2)
```

```python
print(books.BookTitle[0], ' vs ', books.BookTitle[1])
```

```python
#https://www.geeksforgeeks.org/python-measure-similarity-between-two-sentences-using-cosine-similarity/

def compareText(X,Y):
    # tokenization 
    X_list = word_tokenize(X)  
    Y_list = word_tokenize(Y) 

    # sw contains the list of stopwords 
    sw = stopwords.words('english')  
    l1 =[];l2 =[] 

    # remove stop words from the string 
    X_set = {w for w in X_list if not w in sw}  
    Y_set = {w for w in Y_list if not w in sw} 

    # form a set containing keywords of both strings  
    rvector = X_set.union(Y_set)  
    for w in rvector: 
        if w in X_set: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in Y_set: l2.append(1) 
        else: l2.append(0) 
    c = 0

    # cosine formula  
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i] 
    cosine = c / float((sum(l1)*sum(l2))**0.5) 
    return cosine
```

```python
compareText(books.BookTitle[0], books.BookTitle[1])
```

```python
books['BookTitle'][0]
```

```python
for bookIndx in range(0,books.BookTitle.size):
    for compareIndx in range(1,books.columns.size):
        compareBook = books.columns[compareIndx]
        #print(books.BookTitle[bookIndx], ' vs ', compareBook)
        cosineSimilarity = compareText(books.BookTitle[bookIndx], compareBook)
        books[compareBook][bookIndx] = cosineSimilarity
```

c. Which two titles are the most similar to each other? Which are the most dissimilar? Where do they rank, among the first 24 results?

```python
dissimilar = pd.DataFrame(books.BookTitle)
dissimilar['tally'] = 0
dissimilar.head(1)
```

```python
mostSimilarCosine = 0
titles = []
for bookIndx in range(0,books.BookTitle.size):
    for compareIndx in range(1,books.columns.size):
        compareBook = books.columns[compareIndx]
        thisCosine = books[compareBook][bookIndx]
        if thisCosine == 1:
            None
        elif thisCosine == 0:
            currentTally = dissimilar.loc[bookIndx, ('tally')]
            dissimilar.loc[bookIndx, ('tally')] = currentTally + 1
        elif thisCosine > mostSimilarCosine:
            mostSimilarCosine = thisCosine
            titles = [books.BookTitle[bookIndx], compareBook]
```

```python
%store mostSimilarCosine
%store titles
%store dissimilar
```

```python
print(mostSimilarCosine)
print(titles)
print('\n\n Most Dissimilar (all valued at 0)')
dissimilar.sort_values('tally',ascending=False)
```

The Cat in the Hat and Green Eggs and Ham were the two most similar. This was largely because their text was contained keywords for a series of similar books from the publisher 'Beginner Books(R) (30 Books)'

The two most dissimilar to the others were the two single word title books: 'Untamed' and 'Later'. These two only had one chance to match up with any of the other titles. It's not surprising to find them the most divergent from the rest.
